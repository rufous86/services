```python
path = '/content/drive/MyDrive/Study/hackathons/datasecrets_first_cup/'
```

```python
df_train = pd.read_csv(path + 'data/train_target.csv')
df_train
```

```python
df_orders = pd.read_csv(path + 'data/orders.csv')
df_orders
```

```python
print('df_train ClientUUId & df_orders ClientUUId\t', len(set(df_train['ClientUUId']) & set(df_orders['ClientUUId'])))
print('df_train ClientUUId\t\t\t\t', len(set(df_train['ClientUUId'])))
print('df_orders ClientUUId\t\t\t\t', len(set(df_orders['ClientUUId'])))
```

```python
df_merge = pd.merge(df_orders, df_train, how='outer', on='ClientUUId')
print(df_merge[['apply_promo_x', 'apply_promo_y']].isna().mean())
df_merge
```

```python
df_clients_promo_october = pd.read_csv(path + 'data/clients_promo_october.csv')
df_clients_promo_october
```

```python
cols = ['OrderUUId', 'addressId', 'deliverySectorId', 'ClientUUId', 'Date',
       'SaleDate', 'UnitUUId', 'NewClient', 'ClientOrderNumber', 'ProductUUId',
       'CategoryId', 'ProductTotalPrice', 'MenuPrice', 'OrderState',
       'OrderPaymentType', 'OrderTotalPrice', 'OrderType_x', 'apply_promo_x']

cols2rename = {'OrderType_x': 'OrderType',
               'apply_promo_x': 'apply_promo'}

df_merge_promo = pd.merge(df_merge, df_clients_promo_october, how='outer', on='ClientUUId')[cols].rename(columns=cols2rename)
```

```python
df_merge_promo = df_merge_promo.dropna(subset=['apply_promo', 'addressId', 'deliverySectorId'])
df_merge_promo
```
