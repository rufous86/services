```python
def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:
    employee = employee.drop_duplicates(subset='salary')
    if employee.shape[0] >= N and N > 0:
        answer = employee['salary'].nlargest(N).iloc[N-1]
        return pd.DataFrame(data={f'getNthHighestSalary({N})': [answer]}, index=[0])
    return pd.DataFrame(data={f'getNthHighestSalary({N})': [None]}, index=[0])

display(nth_highest_salary(employee1, 2))
display(nth_highest_salary(employee2, 2))
display(nth_highest_salary(employee1, 1))
display(nth_highest_salary(employee3, 88))
display(nth_highest_salary(employee1, -1))
```

| Column Name | Type |

| id          | int  |
| salary      | int  |

id is the primary key (column with unique values) for this table.

Each row of this table contains information about the salary of an employee.
 
Write a solution to find the second highest salary from the Employee table. If there is no second highest salary, return null (return None in Pandas).

The result format is in the following example.

Example 1:

Input: 

Employee table:

| id | salary |
|:---:|:---:|
| 1  | 100    |
| 2  | 200    |
| 3  | 300    |

Output: 

| SecondHighestSalary |
|:---:|
| 200                 |

Example 2:

Input: 

Employee table:

| id | salary |
|:---:|:---:|
| 1  | 100    |

Output: 

| SecondHighestSalary |
|:---:|
| null                |

```python
import pandas as pd

data = {
    'id': [1, 2, 3],
    'salary': [100, 200, 300]
}

employee1 = pd.DataFrame(data)
display(employee1)

data = {
    'id': [1],
    'salary': [100]
}

employee2 = pd.DataFrame(data)
display(employee2)
```

```python
def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:
    employee = employee.drop_duplicates(subset='salary')
    if employee.shape[0] >= 2 and 2 > 0:
        answer = employee['salary'].nlargest(2).iloc[1]
        return pd.DataFrame(data={'SecondHighestSalary': [answer]})
    return pd.DataFrame(data={'SecondHighestSalary': [None]})

display(second_highest_salary(employee1))
display(second_highest_salary(employee2))
```
